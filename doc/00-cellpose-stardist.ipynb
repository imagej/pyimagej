```python
# Import required libraries for Cellpose/StarDist segmentation
import imagej.convert as convert
import numpy as np
import matplotlib.pyplot as plt
from cellpose import models
from csbdeep.utils import normalize
from stardist.models import StarDist2D

print("Libraries imported successfully!")
print(f"Using PyImageJ version: {ij.getVersion()}")
```

```python
def filter_index_image(narr: np.ndarray, min_size: int, max_size: int):
    """
    Filter an index image's labels based on a range of minimum and maximum values
    that define the number of pixels that make up a given label.
    """
    unique = np.unique(narr)
    for label in unique:
        if label == 0:
            # skip the background
            continue

        # create a crop for each label
        bbox = get_bounding_box(np.where(narr == label))
        bbox_crop = narr[bbox[0]:bbox[2] + 1, bbox[1]:bbox[3] + 1].copy()
        bbox_crop[bbox_crop != label] = 0

        # get the number of pixels in label
        bbox_crop = bbox_crop.astype(bool)
        label_size = np.sum(bbox_crop)

        if not min_size <= label_size <= max_size:
            narr[narr == label] = 0

    return narr

def get_bounding_box(indices: np.ndarray):
    """
    Get the bounding box coordinates from the label indices.
    """
    # get min and max bounds of indices array
    min_row = np.min(indices[0])
    min_col = np.min(indices[1])
    max_row = np.max(indices[0])
    max_col = np.max(indices[1])

    return (min_row, min_col, max_row, max_col)

print("Utility functions defined!")
```

# Cellpose/StarDist Segmentation Example

Adapted from https://py.imagej.net/en/latest/Cellpose-StarDist-Segmentation.html

In this section, we'll use PyImageJ with Cellpose and StarDist to segment HeLa cell cytoplasms and nuclei. We'll work with a 3D dataset (X, Y, Channel) of HeLa cells expressing a YFP tagged APOBEC3G (A3G) protein with a nuclear stain.

The dataset has two channels:
1. **DAPI** (nuclear stain) - we'll use StarDist for nuclear segmentation
2. **YFP** (cytoplasmic A3G) - we'll use Cellpose for cytoplasm segmentation

Let's start by loading and displaying the image data.

```python
# Load the HeLa cell image data using PyImageJ
data = ij.io().open('https://media.imagej.net/pyimagej/3d/hela_a3g.tif')
xdata = ij.py.from_java(data)

print(f"Image shape: {xdata.shape}")
print(f"Image data type: {type(xdata)}")

# Create a matplotlib subplot to display both channels
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Display each channel (image dimensions are: row, col, ch)
axes[0].imshow(xdata[:, :, 0], cmap="gray")
axes[0].set_title('Nuclear Channel (DAPI)')
axes[0].axis('off')

# Enhance the cytoplasmic channel for better visibility
axes[1].imshow(xdata[:, :, 1] * 125, cmap="gray")
axes[1].set_title('Cytoplasmic Channel (YFP)')
axes[1].axis('off')

plt.tight_layout()
plt.show()
```

**Note:** The cytoplasmic channel is quite dark in the original image, so we multiplied it by 125 for better visualization. This only affects the display and doesn't alter the original data that we'll use for segmentation.

## Running StarDist and Cellpose

Now let's run the segmentation algorithms:
- **StarDist** on the nuclear channel (DAPI)
- **Cellpose** on the cytoplasmic channel (YFP)

```python
# Run StarDist on the nuclear channel
print("Loading StarDist model and segmenting nuclei...")
stardist_model = StarDist2D.from_pretrained('2D_versatile_fluo')
nuc_labels, _ = stardist_model.predict_instances(normalize(xdata[:, :, 0]))

# Run Cellpose on the cytoplasmic channel
print("Loading Cellpose model and segmenting cytoplasm...")
cellpose_model = models.CellposeModel(gpu=False, model_type='cyto')
ch = [0, 0]  # grayscale image
cyto_labels = cellpose_model.eval(xdata[:, :, 1].data, channels=ch, diameter=72.1)

print("Segmentation complete!")
print(f"StarDist found {len(np.unique(nuc_labels))-1} nuclear objects")
print(f"Cellpose found {len(np.unique(cyto_labels[0]))-1} cytoplasmic objects")
```

```python
# Display StarDist results
fig_n, axes_n = plt.subplots(1, 2, figsize=(12, 6))
axes_n[0].imshow(xdata[:, :, 0], cmap="gray")
axes_n[0].set_title('Nuclear Channel Input')
axes_n[0].axis('off')

axes_n[1].imshow(nuc_labels, cmap="viridis")
axes_n[1].set_title('StarDist Nuclear Segmentation')
axes_n[1].axis('off')

plt.tight_layout()
plt.show()

# Display Cellpose results
fig_c, axes_c = plt.subplots(1, 2, figsize=(12, 6))
axes_c[0].imshow(xdata[:, :, 1] * 125, cmap="gray")
axes_c[0].set_title('Cytoplasmic Channel Input')
axes_c[0].axis('off')

axes_c[1].imshow(cyto_labels[0], cmap="cividis")
axes_c[1].set_title('Cellpose Cytoplasm Segmentation')
axes_c[1].axis('off')

plt.tight_layout()
plt.show()
```

## Filtering Segmentation Results

The nuclear segmentation includes some small debris and artifacts that we don't want. Let's filter out objects that are too small or too large using our utility function.

```python
# Filter small debris and too-large objects from nuclear segmentation
# Objects between 500 and 10,000 pixels are likely to be real nuclei
filtered_nuc_labels = filter_index_image(nuc_labels.copy(), 500, 10000)

print(f"Before filtering: {len(np.unique(nuc_labels))-1} objects")
print(f"After filtering: {len(np.unique(filtered_nuc_labels))-1} objects")

# Display the filtered results
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].imshow(xdata[:, :, 0], cmap="gray")
axes[0].set_title('Original Nuclear Channel')
axes[0].axis('off')

axes[1].imshow(nuc_labels, cmap="viridis")
axes[1].set_title('StarDist (Unfiltered)')
axes[1].axis('off')

axes[2].imshow(filtered_nuc_labels, cmap="viridis")
axes[2].set_title('StarDist (Filtered)')
axes[2].axis('off')

plt.tight_layout()
plt.show()
```

## Converting to PyImageJ ROIs

Now let's convert our segmentation results into PyImageJ-compatible ROIs that can be used for further analysis. We'll create both ImgLib2 ROIs (for headless analysis) and traditional ImageJ ROIs.

```python
# Convert segmentation results to ImgLib2 ROI trees
nuc_roi_tree = convert.index_img_to_roi_tree(ij, filtered_nuc_labels)
cyto_roi_tree = convert.index_img_to_roi_tree(ij, cyto_labels[0])

print(f"Nuclear ROIs created: {len(nuc_roi_tree.children())}")
print(f"Cytoplasmic ROIs created: {len(cyto_roi_tree.children())}")

# Display information about the first few nuclear ROIs
print("\nFirst few nuclear ROIs:")
for i in range(min(3, len(nuc_roi_tree.children()))):
    roi = nuc_roi_tree.children().get(i).data()
    print(f"  ROI {i+1}: {type(roi).__name__}")

# Example: Convert an ImgLib2 ROI to a legacy ImageJ ROI
if len(nuc_roi_tree.children()) > 0:
    import scyjava as sj

    imglib_roi = nuc_roi_tree.children().get(0).data()
    ij_roi = ij.convert().convert(imglib_roi, sj.jimport('ij.gui.PolygonRoi'))
    print(f"\nConverted first nuclear ROI to: {type(ij_roi)}")
```

```python
# Final comparison: Display both segmentation results side by side
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

axes[0].imshow(filtered_nuc_labels, cmap="viridis")
axes[0].set_title(f'Nuclear Segmentation\n({len(np.unique(filtered_nuc_labels))-1} objects)')
axes[0].axis('off')

axes[1].imshow(cyto_labels[0], cmap="cividis")
axes[1].set_title(f'Cytoplasm Segmentation\n({len(np.unique(cyto_labels[0]))-1} objects)')
axes[1].axis('off')

plt.tight_layout()
plt.show()

print("🎉 Cellpose/StarDist segmentation complete!")
print("\nNext steps you could explore:")
print("- Measure properties of segmented objects using PyImageJ")
print("- Export ROIs for use in other ImageJ workflows")
print("- Apply additional filtering based on shape or intensity")
print("- Combine nuclear and cytoplasmic segmentations for analysis")
```
